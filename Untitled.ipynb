```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
from keras.models import Sequential
from keras.layers import LSTM, Dense
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from statsmodels.tsa.vector_ar.var_model import VAR
from sklearn.preprocessing import StandardScaler

# Instantiate the scaler
scaler = StandardScaler()
df = pd.DataFrame(np.column_stack((X_inicial, y_inicial)), columns=['X_inicial_0', 'X_inicial_1', 'X_inicial_2', 'X_inicial_3', 'X_inicial_4', 'y_inicial'])

scaled_values = scaler.fit_transform(df)
print(df)

df_scaled = pd.DataFrame(scaled_values, 
                         columns=df.columns, 
                         index=df.index)
print(df_scaled)
# Insantiate VAR model
model = VAR(df_scaled)
```

```python
optimal_lags = model.select_order()

print(f"The optimal lag order selected: {optimal_lags.selected_orders}")
```

```python
lag_order = optimal_lags.selected_orders['bic']
results = model.fit(lag_order)

# Estimate the model (VAR) and show summary
var_model = results.model
results.summary()
```

```python
# Forecast next two weeks
horizon = 2
forecast = results.forecast(df_scaled.values[-lag_order:], steps=horizon)

# Convert to dataframe
df_forecast = pd.DataFrame(forecast, 
                           columns=df_scaled.columns, 
                           index=df_scaled.iloc[:horizon].index)
```

```python
ax = df_scaled[-30:].X_inicial_0.plot(figsize=(10,5))
df_scaled[:horizon].X_inicial_0.plot(ax=ax)
df_forecast.X_inicial_0.plot(ax=ax)
plt.grid(alpha=0.5, which='both')
plt.xlabel('Date')
plt.ylabel('Increment in number of cases')
plt.legend(['Train', 'Test', 'Forecast'])
plt.show()
```
