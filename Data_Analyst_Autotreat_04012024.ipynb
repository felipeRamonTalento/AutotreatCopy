### Técnica de detección de Outliers en las variables numéricas.

```python
# Obtén las columnas numéricas
columnas_numericas = df_autotreat.select_dtypes(include=['float64', 'int64']).columns

# Seis sigma, estas variables al no llevar una distribución normal rapido le llaman la atencion los valores 
# que en principio deberian estar bien.
for col in columnas_numericas:
    print(' ************************* ')
    mu = np.mean(df_autotreat[col])
    sigma = np.std(df_autotreat[col])
    values = df_autotreat[col]
    outliers = ( values < ( mu - 3 * sigma ) ) | ( values > ( mu + 3 * sigma ) )
    print(values[outliers])
```

```python
# Detección de outliers usando el método IQR
outliers_iqr = {}

for col in columnas_numericas:
    Q1 = df_autotreat[col].quantile(0.25)
    Q3 = df_autotreat[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = (df_autotreat[col] < (Q1 - 1.5 * IQR)) | (df_autotreat[col] > (Q3 + 1.5 * IQR))
    outliers_iqr[col] = df_autotreat[col][outliers]

# Imprimir la lista de outliers después de completar el bucle
print(outliers_iqr)
```

```python
# Lista de columnas específicas para imputar
columnas_especificas = ['A'] #'humidity', 'ambient_temperature', 'bath_conductivity'

# Detección de outliers usando el método IQR
outliers_iqr = {}

for col in columnas_especificas:  # Cambio aquí para iterar solo sobre las columnas específicas
    Q1 = df_autotreat[col].quantile(0.25)
    Q3 = df_autotreat[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = (df_autotreat[col] < (Q1 - 1.5 * IQR)) | (df_autotreat[col] > (Q3 + 1.5 * IQR))
    outliers_iqr[col] = outliers

# Imputar outliers en las columnas especificadas utilizando la media
for col, outlier_condition in outliers_iqr.items():
    mean_value = df_autotreat[col].mean()  # Cambio aquí para usar la media
    
    # Imputar outliers con la media
    df_autotreat.loc[outlier_condition, col] = mean_value

# Verificar si los outliers fueron imputados para las columnas especificadas
outliers_after_imputation = {}
for col in columnas_especificas:  # Cambio aquí para verificar solo las columnas específicas
    Q1 = df_autotreat[col].quantile(0.25)
    Q3 = df_autotreat[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = (df_autotreat[col] < (Q1 - 1.5 * IQR)) | (df_autotreat[col] > (Q3 + 1.5 * IQR))
    outliers_after_imputation[col] = df_autotreat[col][outliers]

outliers_after_imputation
```

```python
# Usando la definición de la lista 'variables' que proporcionaste anteriormente

variables = [
    ('V', 'Voltaje', 'green'),
    ('A', 'Intensidad', 'black'),
    ('T', 'Temperatura', 'purple'),
    ('voltage', 'Voltaje', 'orange'),
    ('holding_time', 'Tiempo_espera', 'violet'),
    ('area', 'Area', 'orange'),
    ('modeling_thickness', 'Espesor_modelado', 'green'),
    ('Imax', 'Corriente_máxima', 'blue'),
]

# Generando gráficos individuales con una representación similar al gráfico combinado, pero añadiendo colores y sombreado
fig, axs = plt.subplots(len(variables), 1, sharex=True, figsize=(15, 35))
fig.tight_layout(pad=4.0)

for idx, (column, title, color) in enumerate(variables):
    # Graficando la variable
    axs[idx].plot(df_autotreat.index, df_autotreat[column], linewidth=2, color=color)
    # Rellenando el área bajo la curva
    axs[idx].fill_between(df_autotreat.index, df_autotreat[column], color=color, alpha=0.1)
    axs[idx].set_title(title, fontsize=14)
    axs[idx].tick_params(axis='both', labelsize=14)
    axs[idx].grid(True, which='both', linestyle='--', linewidth=0.5)

axs[-1].set_xlabel('Tiempo', fontsize=16)

plt.show()
```
